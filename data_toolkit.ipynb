3.  Explain the following terms:
1 The difference in np.array, np.asarray and np.asanyarray.
2 The difference between Deep copy and shallow copy.

Answer :
1)

**np.array:**

Creates a new array from the input data.
Always copies the data, unless copy=False is explicitly set and the input is already an array of the desired type.

**np.asarray:**

Converts the input to an array.
Does not copy data if the input is already an ndarray of the correct dtype.

**np.asanyarray:**

Similar to np.asarray, but passes through subclasses of ndarray (e.g., matrix).
Ensures the result is an array or a subclass.

2)

**Shallow Copy:**

Creates a new object, but inserts references to the objects found in the original.

Changes to mutable objects (like lists) in the copied object will affect the original object.

**Deep Copy:**

Creates a new object and recursively copies all objects found in the original.

Changes to any object in the copied structure will not affect the original object.

5.  Create a NumPy array with random integers Petween 1 and 10 of shape (5,6). After creating the array
perform the following operations:

 a)Extract all even integers from array.

 b)Extract all odd integers from array.

```python
import numpy as np

# Create a NumPy array with random integers between 1 and 10 of shape (5,6)
random_array = np.random.randint(1, 11, size=(5, 6))

print("Original Array:")
print(random_array)

# Extract all even integers from array
even_numbers = random_array[random_array % 2 == 0]

print("\nEven Integers:")
print(even_numbers)

# Extract all odd integers from array
odd_numbers = random_array[random_array % 2 != 0]

print("\nOdd Integers:")
print(odd_numbers)
```

6. Create a 3D Numpy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the following operations:

a) Find the indices of the maximum values along each depth level (third axis).

b) Perform element-wise multiplication of between both array.

```python
import numpy as np

# Create a 3D NumPy array of shape (3, 3, 3) with random integers between 1 and 10
random_array = np.random.randint(1, 11, size=(3, 3, 3))

print("Original 3D Array:")
print(random_array)

# Find the indices of the maximum values along each depth level (third axis)
max_indices = np.argmax(random_array, axis=2)

print("\nIndices of Maximum Values along Third Axis:")
print(max_indices)

# Perform element-wise multiplication between the array and its transpose
elementwise_multiplication = random_array * np.transpose(random_array, axes=(0, 2, 1))

print("\nElement-wise Multiplication with Transpose:")
print(elementwise_multiplication)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate a date range
dates = pd.date_range(start='2023-01-01', periods=30, freq='D')

# Generate random data for Temperature and Humidity
np.random.seed(0)  # For reproducibility
temperature = np.random.uniform(low=15, high=30, size=30)
humidity = np.random.uniform(low=40, high=80, size=30)

# Create the DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
})

# Set 'Date' as the index
df.set_index('Date', inplace=True)

print(df.head())

# Create the figure and the first y-axis
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df.index, df['Temperature'], color='tab:red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (Â°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create the second y-axis and plot Humidity
ax2 = ax1.twinx()
ax2.plot(df.index, df['Humidity'], color='tab:blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
fig.tight_layout()  # Adjust layout to accommodate both y-axes
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Generate the Data
data = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Plot the Histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Step 3: Overlay the PDF
# Generate a range of values
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2)

# Step 4: Label the Axes and Set the Title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate Random Data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

# Step 2: Determine Quadrants
def determine_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Quadrant I'
    elif x < 0 and y >= 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})
df['Quadrant'] = df.apply(lambda row: determine_quadrant(row['x'], row['y']), axis=1)

# Step 3: Create the Scatter Plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1', s=100)

# Step 4: Add Labels and Title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Add a legend
plt.legend(title='Quadrant')

# Show the plot
plt.show()
```

```python
from bokeh.plotting import figure, show, output_file
import numpy as np

# Step 1: Generate Data
x = np.linspace(0, 4 * np.pi, 100)  # 100 points from 0 to 4*pi
y = np.sin(x)

# Step 2: Create a Bokeh Figure
p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis')

# Step 3: Add a Line to the Figure
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Step 4: Customize Grid Lines (default grid lines are already present, so this is optional)
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

# Step 5: Output to HTML file and show
output_file("sine_wave.html")
show(p)
```

```python
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6
import numpy as np

# Step 1: Generate Random Categorical Data
categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = np.random.randint(1, 101, size=len(categories))

# Step 2: Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Step 3: Create a Bokeh Figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value', height=400, width=600)

# Step 4: Add a Bar Renderer
p.vbar(x='categories', top='values', width=0.9, source=source,
       legend_field='categories', line_color='white',
       fill_color=factor_cmap('categories', palette=Spectral6, factors=categories))

# Step 5: Add Hover Tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Step 6: Customize Plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Step 7: Output to HTML file and show
output_file("categorical_bar_chart.html")
show(p)
```
