```python
import random
import statistics
import numpy as np
from scipy import stats

int_list = [random.randint(90, 130) for _ in range(100)]

def calculate_mean(lst):
    return statistics.mean(lst)

def calculate_median(lst):
    return statistics.median(lst)

def calculate_mode(lst):
    try:
        return statistics.mode(lst)
    except statistics.StatisticsError:
        return "No unique mode found"

def weighted_mean(values, weights):
    return np.average(values, weights=weights)

def geometric_mean(lst):
    return stats.gmean(lst)

def harmonic_mean(lst):
    return stats.hmean(lst)

def midrange(lst):
    return (min(lst) + max(lst)) / 2

def trimmed_mean(lst, proportion_to_cut):
    return stats.trim_mean(lst, proportion_to_cut)

print(f"Mean: {calculate_mean(int_list)}")
print(f"Median: {calculate_median(int_list)}")
print(f"Mode: {calculate_mode(int_list)}")
print(f"Geometric Mean: {geometric_mean(int_list)}")
print(f"Harmonic Mean: {harmonic_mean(int_list)}")
print(f"Midrange: {midrange(int_list)}")
print(f"Trimmed Mean (10%): {trimmed_mean(int_list, 0.1)}")
```

```python
int_list2 = [random.randint(200, 300) for _ in range(500)]

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.histplot(int_list2, kde=False, color='blue', stat="density", label='Frequency')

sns.kdeplot(int_list2, color='red', label='KDE')

sns.kdeplot(int_list2, color='green', fill=True, label='Gaussian Distribution')
plt.legend()
plt.title('Frequency, Gaussian, and KDE Plot')
plt.show()

def calculate_range(lst):
    return max(lst) - min(lst)

def variance_and_std(lst):
    return statistics.variance(lst), statistics.stdev(lst)

def iqr(lst):
    q1, q3 = np.percentile(lst, [25, 75])
    return q3 - q1

def coefficient_of_variation(lst):
    return np.std(lst) / np.mean(lst)

def mean_absolute_deviation(lst):
    return np.mean(np.abs(lst - np.mean(lst)))

def quartile_deviation(lst):
    q1, q3 = np.percentile(lst, [25, 75])
    return (q3 - q1) / 2

def range_based_dispersion(lst):
    return (max(lst) - min(lst)) / (max(lst) + min(lst))

print(f"Range: {calculate_range(int_list2)}")
var, std = variance_and_std(int_list2)
print(f"Variance: {var}, Standard Deviation: {std}")
print(f"IQR: {iqr(int_list2)}")
print(f"Coefficient of Variation: {coefficient_of_variation(int_list2)}")
print(f"MAD: {mean_absolute_deviation(int_list2)}")
print(f"Quartile Deviation: {quartile_deviation(int_list2)}")
print(f"Range-based Coefficient of Dispersion: {range_based_dispersion(int_list2)}")
```

```python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        mean = self.expected_value()
        return sum((value - mean) ** 2 * prob for value, prob in zip(self.values, self.probabilities))


values = [0, 1, 2, 3, 4]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]
rv = DiscreteRandomVariable(values, probabilities)
print(f"Expected Value: {rv.expected_value()}")
print(f"Variance: {rv.variance()}")
```

```python
import random

def roll_die(n):
    outcomes = [random.randint(1, 6) for _ in range(n)]
    return outcomes

outcomes = roll_die(1000)

expected_value = sum(outcomes) / len(outcomes)
variance = sum((x - expected_value) ** 2 for x in outcomes) / len(outcomes)

print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

tips = sns.load_dataset('tips')

def calculate_skewness(df, column):
    return df[column].skew()

print(f"Skewness of 'total_bill': {calculate_skewness(tips, 'total_bill')}")
print(f"Skewness of 'tip': {calculate_skewness(tips, 'tip')}")

def check_skewness(df, column):
    skewness = df[column].skew()
    if skewness > 0:
        return "Positive Skewness"
    elif skewness < 0:
        return "Negative Skewness"
    else:
        return "Symmetric"

print(f"Skewness of 'total_bill': {check_skewness(tips, 'total_bill')}")
print(f"Skewness of 'tip': {check_skewness(tips, 'tip')}")

def calculate_covariance(df, col1, col2):
    return df[[col1, col2]].cov().iloc[0, 1]

print(f"Covariance between 'total_bill' and 'tip': {calculate_covariance(tips, 'total_bill', 'tip')}")

def calculate_pearson_correlation(df, col1, col2):
    return df[col1].corr(df[col2])

print(f"Pearson correlation between 'total_bill' and 'tip': {calculate_pearson_correlation(tips, 'total_bill', 'tip')}")

import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatterplot of 'total_bill' vs 'tip'")
plt.show()
```

```python
import numpy as np
from scipy import stats

old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases out of 1000 visitors

t_stat, p_value = stats.ttest_ind(new_layout, old_layout)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("The new layout is successful (statistically significant).")
else:
    print("No significant difference between the old and new layouts.")
```

```python
import numpy as np
from scipy import stats

response_times = np.array([143, 38, 51, 4.0, 4.7, 42, 5.2, 4.5, 4.5, 4.4])

# Null hypothesis: Mean response time >= 5
# Alternative hypothesis: Mean response time < 5
population_mean = 5  # Claimed average time
sample_mean = np.mean(response_times)
sample_std = np.std(response_times, ddof=1)  # Sample standard deviation
n = len(response_times)

# Z-test calculation
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = stats.norm.cdf(z_score)  # One-tailed test (left side)

print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("The customer service department's claim is true (statistically significant).")
else:
    print("The customer service department's claim is false.")
```

```python
import numpy as np
from scipy import stats

# Data for click-through rates
layout_a_clicks = np.array([28, 32, 33, 29, 31, 34, 30, 35, 36, 37])
layout_b_clicks = np.array([40, 41, 38, 42, 39, 44, 43, 41, 45, 47])

# Perform independent t-test
t_stat, p_value = stats.ttest_ind(layout_b_clicks, layout_a_clicks)

# Degrees of freedom
df = len(layout_a_clicks) + len(layout_b_clicks) - 2

print(f"T-statistic: {t_stat}")
print(f"Degrees of freedom: {df}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Layout B has a higher click-through rate (statistically significant).")
else:
    print("No significant difference between the two layouts.")
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Generate synthetic data for age groups and voter preferences
np.random.seed(0)  # For reproducibility
age_groups = np.random.choice(['18-30', '31-50', '51+', '18-30', '31-50', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a contingency table
contingency_table = np.array([
    [sum((age_groups == '18-30') & (voter_preferences == 'Candidate A')), sum((age_groups == '18-30') & (voter_preferences == 'Candidate B'))],
    [sum((age_groups == '31-50') & (voter_preferences == 'Candidate A')), sum((age_groups == '31-50') & (voter_preferences == 'Candidate B'))],
    [sum((age_groups == '51+') & (voter_preferences == 'Candidate A')), sum((age_groups == '51+') & (voter_preferences == 'Candidate B'))]
])

# Perform chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square: {chi2}")
print(f"P-value: {p_value}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies: \n", expected)
alpha = 0.05
if p_value < alpha:
    print("There is a significant association between age group and voter preference.")
else:
    print("No significant association between age group and voter preference.")
```
